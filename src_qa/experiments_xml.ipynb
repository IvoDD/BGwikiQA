```python
import xml.etree.ElementTree as ET
tree = ET.parse('bgwiki-20180920-corpus.xml').getroot()
```

```python
cnt = tree[0].find('content')
cnt[2]
```

```python
def innertext(tag):
    return (tag.text or '') + ''.join(innertext(e) for e in tag) + (tag.tail or '')

txt = innertext(cnt[1])
txt
```

```python
cap = "АБВГДЕЖЗИЙКЛМНОПРСТУФХЦЧШЩЬЪЮЯ"

sent = ""
last = ""
for c in txt:
    if not c.isspace():
        if last == '.' and c in cap:
            print(sent)
            sent = ""
        last = c
    sent += c
```
